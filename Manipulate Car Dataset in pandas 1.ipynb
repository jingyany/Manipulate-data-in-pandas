#### Create a dataset of the cars that each includes Manufacturer (For each record, the first word appears in a string at Vehicle name is its manufacturer. For example, Acura 3.5 RL 4dr’s manufacturer is Acura), Vehicle Name, type (i.e., Small/Sporty/Compact/Large Sedan, Sports Car, SUV, Wagon, Minivan, or Pickup), Retail Price, Dealer Cost, City MPG, Hwy MPG.

> The header of this data should sequentially include “manufacturer”, “name”, “type”, “retail_price”, “dealer_cost”, “mpg_city”, and “mpg_hwy”. 



```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
data = pd.read_csv('2004car.csv')
```

```python
data.head()
```

```python
data1 = data['Vehicle Name']
data1.head()
```

```python
manufacturer = data1.str.split('\\s').str[0]
manufacturer = pd.DataFrame(manufacturer)
manufacturer= manufacturer.rename(columns={'Vehicle Name':'Manufacturer'})
manufacturer.head()
```

```python
cartype = data.loc[:,['Small/Sporty/ Compact/Large Sedan', 'Sports Car','SUV', 'Wagon', 'Minivan', 'Pickup']]
cartype['Small/Sporty/ Compact/Large Sedan'] = cartype['Small/Sporty/ Compact/Large Sedan'].map({1: 'Small/Sporty/ Compact/Large Sedan'})
cartype['Sports Car'] = cartype['Sports Car'].map({1: 'Sports Car'})
cartype['SUV'] = cartype['SUV'].map({1: 'SUV'})
cartype['Wagon'] = cartype['Wagon'].map({1: 'Wagon'})
cartype['Minivan'] = cartype['Minivan'].map({1: 'Minivan'})
cartype['Pickup'] = cartype['Pickup'].map({1: 'Pickup'})
cartype['type'] = cartype.max(axis=1)
cartype.head()
```

```python
car1= pd.DataFrame()
car1['manufacturer']= manufacturer['Manufacturer']
car1['name']= data['Vehicle Name']
car1['type']= cartype['type']
car1['retail_price']= data['Retail Price']
car1['dealer_cost']= data['Dealer Cost']
car1['mpg_city']= data['City MPG']
car1['mpg_hwy']= data['Hwy MPG']

car1.head()
```

```python
car1.to_csv('a5_data1_your_net_id.csv')
```
