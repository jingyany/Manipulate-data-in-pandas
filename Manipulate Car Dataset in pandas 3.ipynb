#### Create a dataset that lists every manufacturer appear in the dataset. For each manufacturer, it is required to present the manufacturer name (e.g., Audi, or BMW), the manufacturer’s average net income (the net income for each car can be calculated by Retail Cost - Dealer Cost), the average of dealers’ profit percentage (the profit percentage can be calculated by (Retail Price - Dealer Cost)/Retail Price*100)).

> The header of this data should include “manufacturer”, “avg_net_income”, “avg_profit_percentage”, “retail_price”, “dealer_cost”, “mpg_city”, and “mpg_hwy”.

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
data = pd.read_csv('2004car.csv')
data.head()
```

```python
data1 = data['Vehicle Name']
manufacturer = data1.str.split('\\s').str[0]
manufacturer = pd.DataFrame(manufacturer)
manufacturer= manufacturer.rename(columns={'Vehicle Name':'Manufacturer'})
manufacturer.head()
```

```python
data1 = data.loc[:,['Retail Price','Dealer Cost']]
data1['net_income']= data1['Retail Price'] - data1['Dealer Cost']
data1.head()
```

```python
data1['profit_percentage']= (data1['net_income']/data1['Dealer Cost'])*100
data1.head()
```

```python
car3 = pd.DataFrame()
car3['manufacturer']=manufacturer['Manufacturer']
car3['avg_net_income']=data1['net_income']
car3['avg_profit_percentage']=data1['profit_percentage']
car3['mpg_city'] = data['City MPG']
car3['mpg_hwy'] = data['Hwy MPG']
car3.head()
```

```python
car3.dtypes
```

```python
car3 = car3.convert_objects(convert_numeric=True)
car3.dtypes
```

```python
car3 = car3.groupby(['manufacturer']).mean()
car3
```

```python
car3.to_csv('a5_data3_your_net_id.csv')
```
