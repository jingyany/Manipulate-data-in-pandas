#### Create a dataset that lists the cars that are either SUV, Wagon, Minivan, or Pickup. For each car, you are required to include Vehicle Name, Vehicle type (i.e., SUV, Wagon, Minivan, or Pickup truck), Retail Price, Engine Size, HP, City MPG, Hwy MPG, Weight, and Wheel Base in the dataset. The data specifications are as follows:

> The header (i.e., the name of each column in the dataset) of this data should sequentially include “name”, “type”, “retail_price”, “engine_size”, “hp”, “mpg_city”, “mpg_hwy”, “weight”, and “wheel_base”.


```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
data = pd.read_csv('2004car.csv')
```

```python
data
```

```python
data = data.sort_values(by='Small/Sporty/ Compact/Large Sedan')
```

```python
data = data[:183]
```

```python
data.head()
```

```python
cartype = data.loc[:,['Small/Sporty/ Compact/Large Sedan', 'Sports Car','SUV', 'Wagon', 'Minivan', 'Pickup']]
```

```python
cartype
```

```python
cartype['Small/Sporty/ Compact/Large Sedan'] = cartype['Small/Sporty/ Compact/Large Sedan'].map({1: 'Small/Sporty/ Compact/Large Sedan'})
cartype['Sports Car'] = cartype['Sports Car'].map({1: 'Sports Car'})
cartype['SUV'] = cartype['SUV'].map({1: 'SUV'})
cartype['Wagon'] = cartype['Wagon'].map({1: 'Wagon'})
cartype['Minivan'] = cartype['Minivan'].map({1: 'Minivan'})
cartype['Pickup'] = cartype['Pickup'].map({1: 'Pickup'})
```

```python
cartype['type'] = cartype.max(axis=1)
```

```python
car2 = pd.DataFrame()
car2['name']= data['Vehicle Name']
car2['type']= cartype["type"]
car2['retail_price']= data['Retail Price']
car2['engine_size']= data['Engine Size (l)']
car2['hp']= data['HP']
car2['mpg_city']= data['City MPG']
car2['mpg_hwy']= data['Hwy MPG']
car2['weight']= data['Weight']
car2['wheel_base']= data['Wheel Base']
car2.head()
```
